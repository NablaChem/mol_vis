## ETKDGv3

```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
# Upload a molecule from SMILES

mol = Chem.MolFromSmiles('[H][C@]12CCCC[C@]11CCN(C)[C@H]2CC2=C1C=C(OC)C=C2')
mol
```

```python
# Add H

mol = Chem.AddHs(mol)
mol
```

```python
# Apply ETKDG to generate 3D geometry

AllChem.EmbedMolecule(mol, AllChem.ETKDGv3())
mol
```

```python
Curcumin_SMILES = "COC1=C(C=CC(=C1)/C=C/C(=O)CC(=O)/C=C/C2=CC(=C(C=C2)O)OC)O" #Add SMILES from Curcumin

Curcumin = Chem.MolFromSmiles(Curcumin_SMILES) # Generate the molecule

Curcumin = Chem.AddHs(Curcumin) #Add H

AllChem.EmbedMolecule(Curcumin, AllChem.ETKDGv3()) # Generate the molecule

Curcumin #Show results
```

```python
Aspirin_SMILES = "CC(=O)OC1=CC=CC=C1C(=O)O" #Add SMILES from Curcumin

Aspirin = Chem.MolFromSmiles(Aspirin_SMILES) # Generate the molecule

Aspirin = Chem.AddHs(Aspirin)

AllChem.EmbedMolecule(Aspirin, AllChem.ETKDGv3())

Aspirin
```

```python
Gingerol_SMILES = "O=C(C[C@@H](O)CCCCC)CCc1cc(OC)c(O)cc1"

Gingerol = Chem.MolFromSmiles(Gingerol_SMILES)

Gingerol = Chem.AddHs(Gingerol)

AllChem.EmbedMolecule(Gingerol, AllChem.ETKDGv3())

Gingerol
```

```python
from rdkit.Chem import Draw
```

```python
smiles_list = ["COC1=C(C=CC(=C1)/C=C/C(=O)CC(=O)/C=C/C2=CC(=C(C=C2)O)OC)O", "CC(=O)OC1=CC=CC=C1C(=O)O",
               "O=C(C[C@@H](O)CCCCC)CCc1cc(OC)c(O)cc1"]

mol_list = []

for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol, AllChem. ETKDGv3())
    mol_list.append(mol)
    
img = Draw.MolsToGridImage(mol_list) #not ETKDGv3
img
```

```python
ps = rdDistGeom.ETKDGv3() 
ps.clearConfs= False #The previous conformations formed will not be deleted
```
